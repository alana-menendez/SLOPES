```python
import earthaccess
from xarray.backends.api import open_datatree
```

```python
results = earthaccess.search_datasets(instrument="OCI")
short_names = [i.summary()["short-name"] for i in results]
short_names
```

## AOP

Rrs(w), aot_865, angstrom, avw

```python
results = earthaccess.search_data(short_name="PACE_OCI_L2_AOP_NRT", count=1)
paths = earthaccess.open(results)
```

```python
datatree = open_datatree(paths[0])
datatree["geophysical_data"]
```
